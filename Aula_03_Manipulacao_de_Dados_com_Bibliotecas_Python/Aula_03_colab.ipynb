```python
# Importando bibliotecas:

# Pandas para poder manipular dados.
import pandas as pd

# Plotly Express para gerar gráficos.
import plotly.express as px
```

```python
# Importando Planilha do Excel.
df_principal = pd.read_excel("/content/tabela_de_acoes.xlsx", sheet_name="Principal")

# Mostrando DataFrame.
df_principal
```

```python
# Lendo tabela Total_de_acoes.
df_total_acoes = pd.read_excel("/content/tabela_de_acoes.xlsx", sheet_name="Total_de_acoes")

# Mostrando DataFrame.
df_total_acoes
```

```python
# Lendo tabela Ticker.
df_ticker = pd.read_excel("/content/tabela_de_acoes.xlsx", sheet_name="Ticker")

# Mostrando DataFrame.
df_ticker
```

```python
# Lendo tabela Chatgpt.
df_chatgpt = pd.read_excel("/content/tabela_de_acoes.xlsx", sheet_name="Chatgpt")

# Mostrando DataFrame.
df_chatgpt
```

```python
# Chamando colunas.
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()

# Mostrando DataFrame.
df_principal
```

```python
# Alterando os nomes das colunas.
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data', 'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()

# Mostrando DataFrame.
df_principal
```

```python
# Criando a coluna de valor_inicial.
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_dia_pct'] + 1)

# Mostrando DataFrame.
df_principal
```

```python
# Agurpando dados das tabelas Principal e Total de Ações.
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código', how='left')

# Mostrando DataFrame.
df_principal
```

```python
# Removendo coluna Código.
df_principal = df_principal.drop(columns=['Código'])

# Mostrando DataFrame.
df_principal
```

```python
# Crinado coluna variacao_real.
df_principal['variacao_real'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']

# Mostrando DataFrame.
df_principal
```

```python
# Formatar números flutuantes com duas casas depois da vírgula.
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Formatando a Coluna Qtde. Teórica em números inteiros.
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)

# Mostrando DataFrame.
df_principal
```

```python
# Trasendo a coluna de identificação das ações que Subiram, Desceram e as que se manteram Estáveis.
df_principal['Resultado'] = df_principal['variacao_real'].apply(lambda x: 'Subiu' if x > 0 else('Desceu' if x < 0 else 'Estável'))

# Mostrar DataFrame
df_principal
```

```python
# Agrupando Tabelas: Principal com Tiker.
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')

# Removendo coluna
df_principal = df_principal.drop(columns=['Ticker'])

# Mostrando DataFrame
df_principal
```

```python
# Agrupando Tabelas: Principal com ChatGPT.
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')

# Removendo coluna
df_principal = df_principal.drop(columns=['Nome da empresa'])

# Mostrando DataFrame
df_principal
```

```python
# Alterando os nomes das colunas.
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtde_ativos', 'Resultado':'status', 'Nome':'nome', 'Segmento':'segmento', 'Idade (anos)':'idade'}).copy()

# Mostrando DataFrame.
df_principal
```

```python
# Trasendo a coluna de categoria de idade.
df_principal['Cat_idade'] = df_principal['idade'].apply(lambda x: 'Mais de 100' if x > 0 else('Menos de 50' if x < 0 else 'Entre 50 e 100'))

# Mostrar DataFrame
df_principal
```

```python
# Alterando os nomes das colunas.
df_principal = df_principal.rename(columns={'Cat_idade':'cat_idade'}).copy()

# Mostrando DataFrame.
df_principal
```

```python
# Calculando o maior valor.
maior = df_principal['variacao_real'].max()

# Calculando o menor valor.
menor = df_principal['variacao_real'].min()

# Calculando a média.
media = df_principal['variacao_real'].mean()

# Calculando a média de quem Subiu.
media_subiu = df_principal[df_principal['status'] == 'Subiu']['variacao_real'].mean()

# Calculando a média de quem Desceu.
media_desceu = df_principal[df_principal['status'] == 'Desceu']['variacao_real'].mean()

# Impressão dos resultados.
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
# Trazendo apenas os que subiram.
df_principal_subiu = df_principal[df_principal['status'] == 'Subiu']

# Mostrando DataFrame.
df_principal_subiu
```

```python
# Agrupando por segmento e somando as variações.
df_principal_segmento = df_principal_subiu.groupby('segmento')['variacao_real'].sum().reset_index()

# Mostrando DataFrame.
df_principal_segmento
```

```python
# Agrupando por Saldo.
df_analise_saldo = df_principal.groupby('status')['variacao_real'].sum().reset_index()

# Mostrando DataFrame.
df_analise_saldo
```

```python
# Criando gráfico de Análise de saldo.
fig = px.bar(df_analise_saldo, x='status', y='variacao_real', text='variacao_real', text_auto='.2s', title='Variação em reais(R$) por Status', hover_data={'variacao_real': ':.2f'})

# Mostrando o gráfico.
fig.show()
```

```python
# Criando gráfico de Variação por segmento.
fig = px.pie(df_principal_segmento, values='variacao_real', names='segmento', title='Variação em reais(R$) por Segmento', labels={'variacao_real': 'Variação em reais(R$)', 'segmento': 'Segmento'}, hover_data={'variacao_real': ':.2f'})

# Mostrando o gráfico.
fig.show()
```
